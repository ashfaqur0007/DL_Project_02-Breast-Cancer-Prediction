```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#cpu or gpu selection is correct or not
device=torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
```

```python
#load data
data=load_breast_cancer()
x=data.data
y=data.target
print("x shape=",x.shape)
print("x shape=",y.shape)
print("y dimension=",y.ndim)
```

```python
print(y)
```

```python
#data split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
print(xtrain.shape,xtest.shape,ytrain.shape,ytest.shape)
```

```python
scaler=StandardScaler()
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
print("xtrain:\n\n",xtrain)
print("xtest\n\n:",xtest)
```

```python
#data=====>pytorch====>gpu
#xtrain====>torch.tensor(1 argument,dtype=torch.float32)====>device
xtrain=torch.tensor(xtrain,dtype=torch.float32).to(device)
ytrain=torch.tensor(ytrain,dtype=torch.float32).to(device)
xtest=torch.tensor(xtest,dtype=torch.float32).to(device)
ytest=torch.tensor(ytest,dtype=torch.float32).to(device)
```

```python
xtrain.shape
```

```python
#nn archi


class NeuralNet(nn.Module):

    def __init__(self, input_size, hidden_size, output_size):
        super(NeuralNet, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        out = self.sigmoid(out)
        return out


```

```python
input_size=xtrain.shape[1]
hidden_size=64 #2^n numbers
output_size=1
learning_rate=0.001
epochs=100
```

```python
#nn to gpu
model = NeuralNet(input_size, hidden_size, output_size).to(device)
```

```python
criterion=nn.BCELoss()
optimizer=optim.Adam(model.parameters(),lr=learning_rate)
```

```python
#train model
for epoch in range(epochs):
  model.train()
  optimizer.zero_grad()
  outputs=model(xtrain)
  loss=criterion(outputs,ytrain.view(-1,1))
  loss.backward()
  optimizer.step()

  with torch.no_grad():
    predicted = outputs.round()
    correct = (predicted == ytrain.view(-1,1)).float().sum()
    accuracy = correct / ytrain.size(0)#now of row sample
  if (epoch+1) % 10 == 0:
    print(f"Epoch [{epoch+1}/{epochs}], Loss : {loss.item():.4f}, Accuracy: {accuracy.item() * 100:.2f}%")#item figure out num and can modify



```

```python
#train

model.eval()
with torch.no_grad():
  outputs=model(xtrain)
  predicted=outputs.round()
  correct=(predicted==ytrain.view(-1,1)).float().sum()
  accuracy=correct/ytrain.size(0)
  print(accuracy.item()*100)
```

```python
#test
model.eval()
with torch.no_grad():
  outputs=model(xtest)
  predicted=outputs.round()
  correct=(predicted==ytest.view(-1,1)).float().sum()
  accuracy=correct/ytest.size(0)
  print(accuracy.item()*100)
```
